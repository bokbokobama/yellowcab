```python
# source: https://www1.nyc.gov/site/tlc/about/tlc-trip-record-data.page
import pandas as pd
from tqdm import tqdm
df = pd.DataFrame()

# Creates a list of every month from January 2009 to June 2022, in a format which mathes the data download link.
minMonth = '2018-01'
maxMonth = '2019-01'
urlList = pd.date_range(minMonth, maxMonth, freq='MS').strftime("%Y-%m").tolist() # min '2009-01', max '2022-06', default '2022-01', '2022-06'

# Creates a dataframe by combining the datasets of all selected months
for month in tqdm(range(len(urlList[:-1]))):
  df = df.append(pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_' + urlList[month] + '.parquet',
                                 columns = ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_distance', 'fare_amount', 'tip_amount', 'total_amount']),
                                 ignore_index = True)
  pass
print('Successfully created dataframe')
df
```

```python
df["trip_time"] = (df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"])
df["average_speed"] = ((3600 * df["trip_distance"]) / (df["trip_time"].dt.total_seconds()))
df['trip_date'] = df['tpep_pickup_datetime'].map(lambda dt: dt.strftime('%Y-%m-%d'))
df
```

```python
grouped_df = df[ (df['trip_date'] >= minMonth) & (df['trip_date'] < maxMonth) ].groupby('trip_date').size().to_frame("count").reset_index()
grouped_df.plot(kind='line', x='trip_date', y='count')
```
